```python
import pandas as pd
import numpy as np
import spacy
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
```

```python
df["text"] = df["title"] + " " + df["body"] + " " + df["labels"]
```

```python
print("🔄 Running Lemmatization...")
nlp = spacy.load("en_core_web_sm")
```

```python
def lemmatize_text(text):
    doc = nlp(text.lower())
    return " ".join([token.lemma_ for token in doc if not token.is_stop and token.is_alpha])
```

```python
df["text_clean"] = df["text"].apply(lemmatize_text)
```

```python
df
```

```python
# === 4. One-Hot Encode Priority Labels ===
print("🧠 One-hot encoding labels...")
ohe = OneHotEncoder(sparse_output=False)
y_ohe = ohe.fit_transform(df[["priority"]])
```

```python
X_train, X_test, y_train_ohe, y_test_ohe = train_test_split(
    df["text_clean"], y_ohe, test_size=0.2, random_state=42
)
```

```python
print("📚 Vectorizing text using TF-IDF...")
vectorizer = TfidfVectorizer(max_features=5000)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
print("🚀 Training XGBoost classifiers...")
models = []
for i in range(y_train_ohe.shape[1]):
    clf = XGBClassifier(use_label_encoder=False, eval_metric='logloss', verbosity=0)
    clf.fit(X_train_vec, y_train_ohe[:, i])
    models.append(clf)
```

```python
print("🔍 Making predictions...")
y_pred_probs = np.column_stack([model.predict_proba(X_test_vec)[:, 1] for model in models])
y_pred_ohe = (y_pred_probs == y_pred_probs.max(axis=1, keepdims=True)).astype(int)
```

```python
# === 9. Decode Predictions and Show Report ===
y_test_labels = ohe.inverse_transform(y_test_ohe)
y_pred_labels = ohe.inverse_transform(y_pred_ohe)
```

```python

print("\n📊 Classification Report:")
print(classification_report(y_test_labels, y_pred_labels))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import f1_score, classification_report
from xgboost import XGBClassifier

# === 1. One-Hot Encode Target ===
print("🧠 Encoding target labels...")
ohe = OneHotEncoder(sparse_output=False)
y_ohe = ohe.fit_transform(df[["priority"]])
classes = ohe.categories_[0]

# === 2. Compute Class Weights ===
print("⚖️ Computing class weights...")
class_counts = df["priority"].value_counts()
total = len(df)
class_weights = {label: total / count for label, count in class_counts.items()}
print("Class Weights:", class_weights)

# === 3. Train/Test Split ===
X_train, X_test, y_train_ohe, y_test_ohe = train_test_split(
    df["text_clean"], y_ohe, test_size=0.2, random_state=42
)

# === 4. TF-IDF Vectorization ===
print("📚 TF-IDF vectorization...")
vectorizer = TfidfVectorizer(max_features=5000)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# === 5. Train XGBoost Models with Class Weights ===
print("🚀 Training XGBoost models...")
models = []

for i in range(y_train_ohe.shape[1]):
    label = classes[i]
    weight = class_weights[label]

    model = XGBClassifier(
        use_label_encoder=False,
        eval_metric='logloss',
        scale_pos_weight=weight,
        max_depth=5,
        learning_rate=0.05,
        n_estimators=300,
        subsample=0.8,
        colsample_bytree=0.8,
        verbosity=0
    )

    model.fit(X_train_vec, y_train_ohe[:, i])
    models.append(model)

# === 6. Underfitting / Overfitting Check ===
print("\n🔍 Checking underfit/overfit:")
for i, model in enumerate(models):
    y_train_pred = model.predict(X_train_vec)
    y_test_pred = model.predict(X_test_vec)

    train_f1 = f1_score(y_train_ohe[:, i], y_train_pred)
    test_f1 = f1_score(y_test_ohe[:, i], y_test_pred)

    print(f"Class: {classes[i]}")
    print(f"  🏋️‍♂️ Train F1 Score: {train_f1:.4f}")
    print(f"  🧪 Test  F1 Score: {test_f1:.4f}\n")

# === 7. Predict and Evaluate Final Performance ===
print("📊 Final Evaluation:")
y_pred_probs = np.column_stack([
    model.predict_proba(X_test_vec)[:, 1] for model in models
])
y_pred_ohe = (y_pred_probs == y_pred_probs.max(axis=1, keepdims=True)).astype(int)

y_test_labels = ohe.inverse_transform(y_test_ohe)
y_pred_labels = ohe.inverse_transform(y_pred_ohe)

print(classification_report(y_test_labels, y_pred_labels))
```
